```python
from ultralytics import YOLO
import math
import cv2
import cvzone
```

```python

#Gọi model đã train và định nghĩa class
model = YOLO("best.pt")
classname = ['fire']

cap = cv2.VideoCapture(0) #index 0 là Mở webcam thiết bị đang dùng
                          #các thiết bị ngoại vi khác nằm ở index 1 -> 9  
                          #có thể sử dụng 
                          #          cap = cv2.VideoCapture("Tên file muốn dự đoán có đuôi .mp4") để dự đoán video có sẵn
if not cap.isOpened():
    print("Không thể mở camera.")
    exit()

while True:
    ret, frame = cap.read()
    result = model(frame)

    #In ra hình bao
    for info in result:
        boxes = info.boxes
        for box in boxes:
            confidence = box.conf[0]
            confidence = math.ceil(confidence * 100)
            Class = int(box.cls[0])

            #Độ chính xác trên 25% thử phán đoán xem có phải lửa không
            if confidence > 25:
                x1,y1,x2,y2 = box.xyxy[0]
                x1, y1, x2, y2 = int(x1),int(y1),int(x2),int(y2)
                cv2.rectangle(frame,(x1,y1),(x2,y2),(0,0,255),5)
                cvzone.putTextRect(frame, f'{classname[Class]} {confidence}%', [x1 + 8, y1 + 100],scale = 1.5, thickness = 2)
    
    cv2.imshow("Fire",frame)
    cv2.waitKey(1)
```
